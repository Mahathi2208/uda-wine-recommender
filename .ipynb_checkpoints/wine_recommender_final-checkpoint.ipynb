### Cosine Similarity

```python
def clean(doc):
    """
    preprocessing of raw reviews
    """
    review = ""
    for word in doc:
        word = re.sub(r'[^\w]', ' ', word)
        review += word 
    return review

lemmer = nltk.stem.WordNetLemmatizer()
def LemTokens(tokens):
    return [lemmer.lemmatize(token) for token in tokens]
remove_punct_dict = dict((ord(punct), None) for punct in string.punctuation)
def LemNormalize(text):
    return LemTokens(nltk.word_tokenize(text.lower().translate(remove_punct_dict)))
```

```python
def get_cosine_simialrity_df(data: pd.DataFrame, user_preferred_attr: str):
    """
    returns the cosine simialrity scores for each review with given
    preferred attributes of wine 
    """
    cleaned_reviews = [clean(user_preferred_attr)] + list(data['reviews'].apply(lambda x: clean(x)).values)
    LemVectorizer = CountVectorizer(tokenizer=LemNormalize, stop_words='english')
    tf_matrix = LemVectorizer.fit_transform(cleaned_reviews).toarray()
    tfidfTran = TfidfTransformer(norm="l2")
    tfidf_matrix = tfidfTran.fit_transform(tf_matrix)
    cos_similarity_matrix = (tfidf_matrix * tfidf_matrix.T).toarray()
    data['tf_idf_cosine_similarity'] = np.vstack(cos_similarity_matrix[0][1:])
    return data.sort_values(by=['tf_idf_cosine_similarity'], ascending=False)
```

```python
#initialize a tokenizer
spacy_tokenizer = nlp.tokenizer
#function to get word embeddings 
def prep(x):
     z=spacy_tokenizer(x) #spacy tokenizer
     #the .vector vectorizes into a word embedding and then reshapes back #to 300 dim
     z=nlp(z).vector.reshape(300,) 
     return z
```

```python
def get_spacy_similarity_df(wine_df: pd.DataFrame, user_preferred_attr: str):
    """
    returns the spacy similarity scores for each review with given
    preferred attributes of wine 
    """
    # getting all the unique products in the data
    list_of_product_names=wine_df['product_name'].unique()
    # given preferred attributes 
    attr_vect = prep(user_preferred_attr)
    spacy_df = wine_df.copy()
    spacy_df['spacy_similarity_score'] = spacy_df['reviews'].apply(lambda y: np.dot(attr_vect, prep(str(y)))/(np.linalg.norm(attr_vect)*np.linalg.norm(prep(str(y)))))
    return spacy_df.sort_values(['spacy_similarity_score'], ascending=[False]) 
```

### TAKS D: Sentiment Analysis

```python
%%time
def get_sentiment(review):
    sid = SentimentIntensityAnalyzer()
    newWords = {'oaky': 2.0, 
                'rich': 2.0, 
                'complex': 2.0,
                'balance': 2.0,
                'angular' : 2.0,
                'backbone' : 2.0,
                'aggressive': -2.0, 
                'acidic': -2.0,
                'alcoholic': -2.0 }
    sid.lexicon.update(newWords)
    score = sid.polarity_scores(review)
    compound = score.get('compound')
    return compound

data['sentiment_score'] = data.apply(lambda x: get_sentiment(x['reviews'].lower()), axis = 1)
```

### Task H - Lift values & Finding most similar brand for given brand 

```python
def get_lift_value(attr, prod, data):
    """
    returns the lift value for given two brands 
    """
    n_attr_prod = 0
    n_attr =0
    i =0
    for comment in list(data['reviews'].values):
        if type(comment) == float:
            i = i+1
            continue
        else:
            word_set = [re.sub(r'[^\w]', ' ', word).strip().lower() for word in comment.split()]
            final_word_set = word_set.copy()
            final_word_set = [list(final_word_set) + [re.sub(r'[^\w]', ' ', w).strip().lower() 
                                                      for w in word.split()] for word in word_set if " " in word]
            final_word_set = set(list(itertools.chain.from_iterable(final_word_set)))
            if set(final_word_set).intersection(set([attr])):
                n_attr+=1
    for comment in list(data[data['product_name']==prod]['reviews'].values):
        if type(comment) == float:
            i = i+1
            continue
        else:
            word_set = [re.sub(r'[^\w]', ' ', word).strip().lower() for word in comment.split()]
            final_word_set = word_set.copy()
            final_word_set = [list(final_word_set) + [re.sub(r'[^\w]', ' ', w).strip().lower() 
                                                      for w in word.split()] for word in word_set if " " in word]
            final_word_set = set(list(itertools.chain.from_iterable(final_word_set)))
            if set(final_word_set).intersection(set([attr])):
                n_attr_prod+=1
    return  (len(data) - i) * (n_attr_prod / (n_attr * len(data[data['product_name']==prod]) ))
```

```python
top_wine_attributes = ['fruit', 'vanilla', 'cherry',  'tannins' ]
wine_brands = list(set(data['product_name'].values))[:10]
# all_list = wine_brands.copy()
# all_list.extend(top_wine_attributes)
len(wine_brands)
```

```python
all_list
```

```python
%%time
attribute_lift_values = []
for brand in wine_brands:
    lift_value_tmp = []
    for attribute in top_wine_attributes:
        lift= get_lift_value(attribute,brand, data)
        lift_value_tmp.append(lift)
    attribute_lift_values.append(lift_value_tmp)
```

```python
attribute_lift_values = pd.DataFrame(attribute_lift_values, columns=top_wine_attributes, index=wine_brands)
attribute_lift_values
```

```python
%%time
cosine_similarity_brands = []
for brand_1 in wine_brands:
    temp = []
    for brand_2 in wine_brands:
        x = np.array(list(attribute_lift_values.loc[brand_1, ].values))
        y = np.array(list(attribute_lift_values.loc[brand_2, ].values))
        temp.append(np.dot(x, y)/(np.linalg.norm(x)*np.linalg.norm(y)))
    cosine_similarity_brands.append(temp)
```

```python
cosine_similarity_brands_df = pd.DataFrame(cosine_similarity_brands, columns=wine_brands, index=wine_brands)
cosine_similarity_brands_df
```

#### Approach taken:
1. Finding lift values for a brand in the randomly selected pool of 10 wine brands with top four attributes of wine 
  i.e; ['fruit', 'vanilla', 'cherry',  'tannins'] (based on word frequency)
2. Creating vector for each brand based on those lift values
3. calculating cosine similarity between the defined vectors for each brand
4. selecting the brand which has high similarity score for the particular brand as the one most similar to it
